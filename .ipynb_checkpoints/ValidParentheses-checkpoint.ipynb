```python
def valid_par(s: str) -> bool:
    #in stack we will place unpaired brackets
    stack = []
    
    #making map for every closing bracket we oppose oppening bracket
    mapping = {')':'(', '}':'{',']':'['}
    
    #going through the string of brackets
    for char in s:
        
        #if element of string in map
        if char in mapping:
            
            #taking out with deleting from the stack the last element if stack is not empty
            top_element = stack.pop() if stack else 0
            
            #if out char from the string is not open bracket to our top element then return False
            if mapping[char] != top_element:
                return False
        #otherwise add char to stack to wait if will come it's own pair of brackets    
        else:
            stack.append(char)
    
    #if stack is not empty in the end ergo any of bracket didn't get it's pair we return notTrue = False
    #and opposite
    return not stack
```

```python
a = "rsdfgdf"
```

```python
valid_par(a)
```
