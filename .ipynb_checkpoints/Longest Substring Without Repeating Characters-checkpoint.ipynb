```python
string_t1 = 'abcabcbb'
string_t2 = 'ffffff'
strinh_t3 = 'pwwkew'
```

```python
def longest_substring(s: str)->int:
    tmp_string = ''
    
    for i in s:
        if i not in tmp_string:
            tmp_string.join(i)
    print(tmp_string)
        
    return len(tmp_string)
```

```python
assert longest_substring(string_t1) == 3
assert longest_substring(string_t2) == 1
assert longest_substring(strinh_t3) == 3
```
