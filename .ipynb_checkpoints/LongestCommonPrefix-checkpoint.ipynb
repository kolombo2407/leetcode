```python
strs1 = ['flower', 'flow', 'flight']
strs2 = ['dog', 'racecar', 'car']
```

```python
def longestPrefix(s: str) -> str:
    #creating future result
    char_set = ''
    #calculating a lengh for the shortest word in list
    smallest_str = min([len(x) for x in s])
    
    #going in lengh for smallest word in list 
    for i in range(smallest_str):
        #taking smallest size batches in every word in list
        co = [x[:smallest_str] for x in s]
        #cutting the lengh of all batches, until the unrepeated set of it will not be equal to 1
        if len(set(co)) == 1:
            return s[0][:smallest_str]
        smallest_str -= 1
    #if reducing the total lengh will bring us to zero return empty string, meaning there are no common prefix
    return ''
```

```python
assert longestPrefix(strs1) == 'fl'
assert longestPrefix(strs2) == ''
```
