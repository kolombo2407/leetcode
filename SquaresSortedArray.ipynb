```python
arr_nums = [[-4,-1,0,3,10], [-7,-3,2,3,11]]
```

```python
def srtd_sq(nums: list) -> int:
    sq_list = result = []
    for num in nums:
        sq_list.append(num**2)
    result = sorted(sq_list)
    return result
```

```python
assert srtd_sq(arr_nums[0]) == [0,1,9,16,100]
assert srtd_sq(arr_nums[1]) == [4,9,9,49,121]
```
