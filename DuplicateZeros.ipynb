```python
arr_list = [[1,0,2,3,0,4,5,0],[1,2,3]]

def duplicateZeros(arr: list)-> None:
    lengh = len(arr)
    result, pos_idx = [], []
    pos = 0

    for a in arr:
        if a == 0:
            pos = arr.index(a, pos)
            pos_idx.append(pos)
            pos +=1

    for p in pos_idx[::-1]:
        arr.insert(p, 0)

    if 0 not in arr:
        arr = arr
    else:
        del arr[lengh-len(arr):]
```

```python
for arr in arr_list:
    duplicateZeros(arr)
    print(arr)
```
