```python
string_t1 = 'abcabcbb'
string_t2 = 'ffffff'
string_t3 = 'pwwkew'
string_t4 = ' '
string_t5 = 'au'
string_t6 = 'dvdf'
```

```python
def longest_substring(s: str)->int:
    chars = [0] * 128

    left = right = 0

    res = 0
    while right < len(s):
        r = s[right]
        chars[ord(r)] += 1

        while chars[ord(r)] > 1:
            l = s[left]
            chars[ord(l)] -= 1
            left += 1

        res = max(res, right - left + 1)

        right += 1
    return res
```

```python
assert longest_substring(string_t1) == 3
assert longest_substring(string_t2) == 1
assert longest_substring(string_t3) == 3
assert longest_substring(string_t4) == 1
assert longest_substring(string_t5) == 2
assert longest_substring(string_t6) == 3
```
