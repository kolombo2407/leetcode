```python
def even_nums(nums: list) -> int:
    count = 0
    for num in nums:
        if len(str(num))%2 == 0:
            count += 1
    return count
            
```

```python
assert even_nums(arr_nums[0]) == 2
assert even_nums(arr_nums[1]) == 1
```
