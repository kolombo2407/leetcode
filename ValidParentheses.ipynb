```python
def valid_par(s: str) -> bool:
    par1 = 0
    par2 = 0
    par3 = 0
    
    for elem in s:
        if elem == '(':
            par1 += 1
        elif elem == ')':
            par1 -= 1
        elif elem == '[':
            par2 += 1
        elif elem == ']':
            par2 -= 1
        elif elem == '{':
            par3 += 1
        elif elem == '}':
            par3 -= 1
        else:
            print ('wrong input')
        
        if par1 < 0 or par2 < 0 or par3 < 0:
            return False
        
        
        
    if par1 + par2 + par3 == 0:
        return True
    else:
        return False
    
    
```

```python
a = "()[]{}"
```

```python
valid_par(a)
```
