```python
containers = [[1,1], [1,8,6,2,5,4,8,3,7], [4,3,6,7,4,6,7,8,8]]
```

```python
def max_container (height: list) -> int:
    '''
    defining starting values for maxarea as 0
    the most left index for elements in height is 0
    the most right index for elements in height is equal of lengh of inputed list minus 1
    '''
    maxarea = 0
    left = 0
    right = len(height) - 1
    
    while left < right:
        # width of the container is defines as difference between indexes of right and left elements
        width = right - left
        
        '''
        calculating area of the container. Taking as a height the smallest between 
        comparing elements and multuply it for width
        '''
        maxarea = max(maxarea, min(height[left], height[right])*width)
        
        #moving towards the list comaring elems        
        if height[left] <= height[right]:
            left += 1
        else:
            right -= 1
            
    return maxarea
```

```python
assert max_container(containers[0]) == 1
assert max_container(containers[1]) == 49
assert max_container(containers[2]) == 36
```
